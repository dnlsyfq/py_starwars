```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("StarWars.csv",encoding="ISO-8859-1")
```

```python
data.head()
```

```python
data = data[data['RespondentID'].notnull()]
```

```python
data.columns
```

```python
data.iloc[:,1].value_counts(dropna=False)
```

```python
data.iloc[:,2].value_counts(dropna=False)
```

```python
true_false = {
    "Yes":True,
    "No":False
}
```

```python
data.iloc[:,1] = data.iloc[:,1].map(true_false)
```

```python
data.iloc[:,2] = data.iloc[:,2].map(true_false)
```

```python
data.head(3)
```

```python
data = data.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.":"seen_1",
    "Unnamed: 4":"seen_2",
    "Unnamed: 5":"seen_3",
    "Unnamed: 6":"seen_4",
    "Unnamed: 7":"seen_5",
    "Unnamed: 8":"seen_6"
})
```

```python
data.head(2)
```

```python
unique_dc = {}
for i in data.columns[3:9]:
    unique_dc[i] = data[i].unique()

unique_dc
```

```python
true_false_film = {
    "Star Wars: Episode I  The Phantom Menace":True,
    "Star Wars: Episode II  Attack of the Clones":True,
    "Star Wars: Episode III  Revenge of the Sith":True,
    "Star Wars: Episode IV  A New Hope":True,
    "Star Wars: Episode V The Empire Strikes Back":True,
    "Star Wars: Episode VI Return of the Jedi":True,
    np.NaN:False,
}
```

```python
for i in data.columns[3:9]:
    data[i] = data[i].map(true_false_film)
```

```python
data[data.columns[3:9]].head()
```

```python
data.head()
```

```python
data.columns[9:15]
```

```python
data[data.columns[9:15]].head()
```

```python
data[data.columns[9:15]] =  data[data.columns[9:15]].astype(float)
```

```python
data[data.columns[9:15]].head()
```

```python
data.rename(columns={
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.":"ranking_1",
    "Unnamed: 10":"ranking_2",
    "Unnamed: 11":"ranking_3",
    "Unnamed: 12":"ranking_4",
    "Unnamed: 13":"ranking_5",
    "Unnamed: 14":"ranking_6"
},inplace=True)
```

1 means the film was the most favorite, and 6 means it was the least favorite. Remember that a lower ranking is better!

```python
data.head()
```

```python
data_avg = data[data.columns[9:15]].mean().reset_index()
```

```python
data_avg.rename(columns={'index':'ranking',0:'avg'},inplace=True)
```

```python
data_avg
```

```python
%matplotlib inli
```

```python
plt.bar(data_avg['ranking'],height=data_avg['avg'])
```

```python
data_sum = data[data.columns[data.columns.str.contains(r'seen_')]].sum().reset_index()
data_sum.rename(columns = {
    'index':'watch',
    0:'Total'
},inplace=True)
```

```python
plt.bar(data_sum['watch'],height=data_sum['Total'])
```
